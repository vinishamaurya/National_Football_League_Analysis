## Creating Dummies

```python
#importing ensemble models
import sklearn.ensemble as ensemble
from sklearn.model_selection import train_test_split
```

```python
#X- y split
X = data[['College wins','College losses','Height (inches)','Month_of_Birth','Hometown','Home State']]
y = data[['Draft Round']]

#creating dummies
cat_vars=['Hometown','Home State']
for var in cat_vars:
    cat_list='var'+'_'+var
    #print(cat_list)
    cat_list = pd.get_dummies(X[var], prefix=var,drop_first=True)
    #print(cat_list)
    X=X.join(cat_list)
    
    X.head()
    #data=data1
X.drop(['Hometown','Home State'],axis=1,inplace=True)
X.head() 

```

```python
#test-train split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

## Q1

### Using Random Forest to predict Draft Round

```python
#calling random forest classifier
randomForest=ensemble.RandomForestClassifier(n_estimators=50,max_features=0.5,oob_score=True)
```

```python
#model fitting on train data
randomForest.fit(X_train,y_train)
```

```python
#checking the base_estimator
randomForest.base_estimator_
```

```python
#predicting on test data
pred2=randomForest.predict(X_test)
pred2[1:5]
```

```python
#checking the class frequency in predicted data
from scipy.stats import itemfreq
itemfreq(pred2)
```

```python
#checking the variable importance
randomForest.feature_importances_     # highest value gives reduction in gini for classification model|||||| for regression -RSS
```

```python
# checking the out of bag score
randomForest.oob_score_
```

```python
#mean accuracy
randomForest.score(X_test,y_test)    #returns the mean accuracy on the given test data and labels
```

```python
#classification report
from sklearn.metrics import classification_report
print(classification_report(y_test, pred2))
```

### Accuracy is very high, that means selected features are able to predict the Draft Round properly. Draft round is dependent on these variables.
